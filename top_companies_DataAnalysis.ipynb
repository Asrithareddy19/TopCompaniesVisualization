```python
import pandas as pd
```

```python
df = pd.read_csv("Top_1000_Companies_Dataset.csv")
df
```

```python
print(df.info())
```

```python
print(df.duplicated().sum())
```

```python
# Drop irrelevant columns
columns_to_drop = ['url', 'indeed_url','linkedin_url','product_url', 'keywords', 'btype','contact_info']
df = df.drop(columns=columns_to_drop, axis=1)
df
```

```python
print(df.info())
```

```python
# Drop columns with more than 50% missing values
df = df.loc[:, df.isnull().mean() < 0.5]
df
```

```python
# Check for missing values in the entire dataset
print(df.isnull().sum())
```

```python
# Calculate the percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage)
```

```python
# Fill missing values for 'estimated_revenues'
df.loc[:, 'estimated_revenues'] = df['estimated_revenues'].fillna(df['estimated_revenues'].median())

# Fill missing values for 'Industry' with mode
df.loc[:, 'Industry'] = df['Industry'].fillna(df['Industry'].mode()[0])

# Fill missing values for 'city' with "Unknown"
df.loc[:, 'city'] = df['city'].fillna("Unknown")
```

```python
# Drop rows where critical columns have missing values
df = df.dropna(subset=['company_name', 'Industry'])
```

```python
# Fill missing values with 0
df['total_funding'] = df['total_funding'].fillna(0)
```

```python
df.info()
```

```python
import re

# Function to clean and convert funding values
def clean_funding(funding):
    funding = str(funding)

    # Remove all non-numeric characters except for digits, periods, 'B', and 'M' (for billion and million)
    funding = re.sub(r'[^\d\.BM]', '', funding)

    # Convert "B" for billion, "M" for million
    if 'B' in funding:
        return float(funding.replace('B', '')) * 1e9
    elif 'M' in funding:
        return float(funding.replace('M', '')) * 1e6
    else:
        return float(funding)  # For numeric values without "B" or "M"

# Apply the cleaning function
df['total_funding'] = df['total_funding'].apply(clean_funding)

print(df['total_funding'].head())
```

```python
# Clean growth_percentage column
df['growth_percentage'] = df['growth_percentage'].replace({',': '', '%': ''}, regex=True).astype(float)
print(df['growth_percentage'].head())
```

```python
# Check for duplicate rows
duplicates = df[df.duplicated()]
print(f"Duplicate rows: {duplicates}")
```

```python
print(df.describe())
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.dropna(inplace=True)
```

```python
print(df.describe())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize total_funding and growth_percentage
scaler = MinMaxScaler()
df[['total_funding', 'growth_percentage']] = scaler.fit_transform(df[['total_funding', 'growth_percentage']])
```
